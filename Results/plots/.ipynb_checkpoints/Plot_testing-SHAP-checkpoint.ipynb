```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score, accuracy_score, make_scorer
import shap
import numpy as np
```

```python
PATH=r"D:\Jupyter\Hard_Metals\240617-Codes_To_publish\ML_Pipeline\240723\\"
```

```python
estimator= pd.read_pickle(f"{PATH}random_forest_model_used.pkl")
explainer= pd.read_pickle(f"{PATH}shap_explainer.pkl")
```

```python
features_trost =  ['slope_0.2',  'norm_slope_0.2',  'curvature','curvature_linear', 'curvature_shift', 'norm_curvature','norm_curvature_linear', 'norm_curvature_shift', 'work_total','work_elastic', 'work_plastic', 'work_ratio', 'h_ratio']
features_others = ['Er(GPa)', 'H(GPa)']
all_features = features_trost.copy()
all_features.extend(features_others)
```

```python
features=all_features
```

```python
PATH=r"D:\Dokumente\General_Data\other_people\Hard_Metalls\240422-new_data\worked_data\Paper\Data_Sets\\"
X_val_df = pd.read_csv(f"{PATH}X_test_240722.csv")
y_val = pd.read_csv(f"{PATH}y_test_240722.csv")
```

```python
y_val["predicted"] = estimator.predict(X_val_df[all_features]) 
```

```python
f1_score(y_val.Y, y_val.predicted, average="weighted")
```

```python
f1_score(y_val.Y, y_val.predicted, average="macro")
```

```python
mapping= {
    'matrix': 0,
    'mc-matrix': 1,
    'm6c-matrix': 2,
    'mc': 3,
    'm6c': 4,
    'mc-mc': 3,
    'm6c-m6c': 4,
    'mc-m6c': 5}
```

```python
labels= {
    'matrix': 0,
    'mc-matrix': 1,
    'm6c-matrix': 2,
    'mc': 3,}
```

```python
cm = confusion_matrix(y_val.Y, y_val.predicted)
labels= {
    r'Matrix': 0,
    r'MC-Matrix': 1,
    r'$\mathrm{M}_6\mathrm{C}$-Matrix': 2,
    r'MC': 3,
    r"$\mathrm{M}_6\mathrm{C}$": 4,}
sns.heatmap(cm, annot=True,xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted Labels', fontweight='bold')
plt.ylabel('True Labels', fontweight='bold')
save_path=r"D:\Dokumente\Paper Writing\Ongoing\Hard_Metalls\images\\"
plt.tight_layout()
#plt.savefig(f"{save_path}testing_confusion_map.jpeg", dpi=600)
```

```python
shap_values = explainer.shap_values(X_val_df[all_features])
```

```python
features_formel={'slope_0.2':r'$\frac{dP}{dH}$',
   'work_total': r'$W$',
 'work_elastic':r'$W_e \text{[}10^{-3}$ µN nm]',
 'work_plastic': r'$W_p$',
 'work_ratio': r'work-ratio',
 'curvature':r"$C_0 \, [\mu \text{N}/\text{nm}^2]$",
 'curvature_linear': r'$C_1$',
 'curvature_shift': r'$C_2$',
 'h_ratio':'h-ratio',
 'norm_curvature': r'$C_{0,norm}$',
 'norm_curvature_linear': r'$C_{1,norm}$',
 'norm_curvature_shift': r'$C_{2,norm}$',
 "norm_slope_0.2": r'${\frac{dP}{dH}}_{norm}$',
                 'Er(GPa)': r'$E_{r}$ [GPa]',
                 'H(GPa)': r'$H$ [GPa]'
                 
                }
```

```python
latex_features=[]
for x in all_features:
    latex_features.append(features_formel[x])
```

```python
labels_reversed= {
    0: r'Matrix',
    1: r'MC-Matrix',
    2: r'$\mathrm{M}_6\mathrm{C}$-Matrix',
    3: r'MC',
    4: r"$\mathrm{M}_6\mathrm{C}$",}
```

```python
sns.set(style="whitegrid")

# Create a figure with subplots: 2 columns for the feature names and 5 rows for each iteration
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))

# List of feature names
names = [r"$C_0 \, [\mu \text{N}/\text{nm}^2]$", r'$H$ [GPa]']

z=0
# Loop through the names and the iterations, then plot SHAP dependence on each subplot
for i in range(5):
    for j, name in enumerate(names):
        axes[i,j].axhline(y=0, xmin=0, xmax=1, linestyle="--", color="grey")
        shap.dependence_plot(
            name, shap_values[i], X_val_df[all_features],
            display_features=None, show=False, ax=axes[i, j], feature_names=latex_features
        )

        # Set gridlines for the current axis
        axes[i, j].grid(True, which='both', linestyle='--', linewidth=0.7)
        axes[i,j].set_title(labels_reversed[i])

        axes[i, j].spines['top'].set_visible(True)
        axes[i, j].spines['right'].set_visible(True)
        axes[i, j].spines['bottom'].set_visible(True)
        axes[i, j].spines['left'].set_visible(True)

        # Add axis lines around the plot
        axes[i, j].spines['top'].set_linewidth(1.5)
        axes[i, j].spines['right'].set_linewidth(1.5)
        axes[i, j].spines['bottom'].set_linewidth(1.5)
        axes[i, j].spines['left'].set_linewidth(1.5)
        axes[i, j].text(0.0, 0.98, f" {chr(ord('a') + z)})", ha='left', va='top', transform=axes[i, j].transAxes, fontsize=20, fontweight='bold')
        z+=1
       # Modify tick labels for the left column only
        if j == 0:  # Left column
            x_ticks_new=np.array(axes[i, j].get_xticks()*10**-6)
            x_ticks_new = np.round(x_ticks_new,2)
            axes[i, j].set_xticklabels(x_ticks_new)
                


        
# Change the colormap
newcmp = plt.get_cmap("viridis")

# Change the colormap of the artists
for fc in plt.gcf().get_children():
    for fcc in fc.get_children():
        if hasattr(fcc, "set_cmap"):
            fcc.set_cmap(newcmp)

# Adjust layout
plt.tight_layout()
save_path = r"D:\Dokumente\Paper Writing\Ongoing\Hard_Metalls\images\\"
plt.savefig(f"{save_path}shap_plot_testing_interactions.jpeg", dpi=600)
```
