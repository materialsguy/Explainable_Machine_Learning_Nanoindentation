# KNN needs to be excluded rom the results dataframe as it was not in a proper pipeline

```python
results_all=results[results.model != "KNN"]
```

```python
all_results= pd.concat([results_all, results_2])
```

# Plot the CV results

```python
fig, ax1 = plt.subplots(1, 1,  dpi=600) 

# Define a vibrant color palette
vibrant_colors = ['#0072B2','#D55E00' ,'#CC79A7']

# Define markers for different models
markers = ['x', '*', '3', "+"]


model_names = {"RFC" : "Random Forest", "KNN": "k-nearest neighbour", "XGB" : "eXtreme Gradient Boosting", "SVC": "Support Vector Classification"}

# Plot by model with custom vibrant colors and markers
for i, model in enumerate([ "XGB", "RFC", "KNN","SVC"]):
    current_data = all_results[all_results['model'] == model]
    for j, num_features in enumerate([2,13,15]):
        plot_data = current_data[current_data.num_features == num_features]
        color =  vibrant_colors[j % len(vibrant_colors)]
        ax1.plot(plot_data['mean_test_f1_weighted'], plot_data['mean_test_f1_macro'], 
                 marker=markers[i % len(markers)], markersize=7, linestyle='', color=color)

ax1.set_xlabel(r'Mean Test $F1_{\text{Weighted}}$', fontsize=16)  # Adjust axis label font size
ax1.set_ylabel(r'Mean Test $F1_{\text{Macro}}$', fontsize=16)
ax1.grid(True, linestyle='--', linewidth=0.5, color='gray', which='both', alpha=0.7)  # Customize gridlines


def create_custom_legend(labels, markers, vibrant_colors, features):
  """Creates a custom legend with grey symbols and custom labels and markers.

  Args:
    labels: A list of strings for the legend labels.
    markers: A list of marker styles for the legend.

  Returns:
    A list of legend handles.
  """

  handles = []
  for i in range(len(labels)):
    handle = plt.plot(0,0, color='grey', marker=markers[i], linestyle='None', markersize=10, label=labels[i])
  for i, feature in enumerate(features):
    handle = plt.plot(0,0, marker='s', color=vibrant_colors[i], linestyle='None', markersize=10, label=feature)
    


# Example usage:
labels = ['Random Forest', 'eXtreme Gradient Boosting',  r"$k$-nearest neighbour", "Support Vector Classification"]
markers = [ '*', 'x', '3', "+"]
features= ["Set A","Set B","Set C"]
handles = create_custom_legend(labels, markers, vibrant_colors, features)

plt.legend(title= "Models and Feature sets", title_fontproperties={'weight':'bold'})
plt.xlim(0.625,0.79)
plt.ylim(0.385,0.54)
save_path=r"D:\Dokumente\Paper Writing\Ongoing\Hard_Metalls\images\\"
plt.tight_layout()
plt.savefig(f"{save_path}cross_validation.jpeg", dpi=600)


```
